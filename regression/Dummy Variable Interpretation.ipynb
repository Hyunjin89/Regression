```python
import numpy as np
import pandas as pd
import statsmodels.api as sm

df = pd.read_csv('house_prices.csv')
df.head()
```

** pandas.get_dummies**

```python
df[['A', 'B', 'C']] = pd.get_dummies(df['neighborhood'])
df[['lodge', 'ranch', 'victorian']] = pd.get_dummies(df['style'])
```

```python
df.head()
```

You always drop one level from each category. 
This is called the baseline

```python
df['intercept'] = 1 

lm = sm.OLS(df['price'], df[['intercept', 'lodge', 'ranch']]) # drop victorian
results = lm.fit()
results.summary()
```

- The intercept means that if our home is a Victorian (Baseline), we predict its price to be 1,046,000 dollars

- A Lodge is predicted to be 741,000 less than a Victorian

- A rnach is predicted to be 471,000 less than a Victorian
